```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mydata=pd.read_csv('mydata.csv')
mydata=mydata.iloc[:,1:]
mydata
```

```python
mydata.columns
```

```python
mydata.dtypes
```

```python
mydata.describe()
```

```python
#find NAs
mydata.isna().sum()
```

```python
#find nulls
mydata.isnull().sum()
```

```python
#get number of wins and losses
mydata.winloss.value_counts()
```

```python
#get number of wins and losses per year
plt.rcParams['figure.figsize']=[12,8]
winloss_by_year=mydata.groupby("year")["winloss"].value_counts()
winloss_by_year.plot(kind="bar",title="Number of Wins and Losses by Year")
```

```python
#number of games by locale by year
plt.rcParams['figure.figsize']=[12,8]
locale_by_year=mydata.groupby("year")["locale"].value_counts()
locale_by_year.plot(kind="bar",title="Number of Home and Away Games by Year")
```

```python
#mean TSU score by year
mean_score_by_year=mydata.groupby("year")["TSU score"].mean()
mean_score_by_year.plot(x="year",y="TSU score",kind="line",title="Mean TSU Score by Year")
```

```python
#descriptive statistics for TSU score by year
mydata.groupby("year")["TSU score"].describe()
```

```python
#mean opponent score by year
mean_oppscore_by_year=mydata.groupby("year")["opponent score"].mean()
mean_oppscore_by_year.plot(x="year",y="opponent score",kind="line",title="Mean Opponent Score by Year")
```

```python
#descriptive statistics for opponenet score
mydata.groupby("year")["opponent score"].describe()
```

```python
#mean absolute value score difference by year
mean_scorediff_by_year=mydata.groupby("year")["scorediff_abs"].mean()
mean_scorediff_by_year.plot(x="year",y="scorediff_abs",kind="line",title="Mean Absolute Value of the Score Difference by Year")
```

```python
#descriptive statistics for absolute difference of score difference
mydata.groupby("year").scorediff_abs.describe()
```

```python
#mean attendance by year
mean_attendance_by_year=mydata.groupby("year")["attendance"].mean()
mean_attendance_by_year.plot(x="year",y="attendance",kind="line",title="Mean Attendance by Year")
```

```python
#descriptive statistics for attendance
mydata.groupby("year").attendance.describe()
```
